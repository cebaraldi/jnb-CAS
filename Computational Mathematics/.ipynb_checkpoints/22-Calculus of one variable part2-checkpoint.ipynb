**Problem:** Consider a function $f(x) = x sin(cos(2x))-exp(-(x-1)^2)$. Plot the graph of $f(x)$ along with first two derivatives in [0.5,2]

```python
f(x)=x*sin(cos(2*x))-exp(-(1-x)^2)
show("f:"+latex(f))
```

```python
f1=f.diff()
show("f':"+latex(f1))
```

```python
f2=f.diff(2)
show("f'':"+latex(f2))
```

Plot the graph of all these three functions:

```python
a,b=-6,6
p=plot(f(x),a,b,color='red',legend_label='$y=f(x)$')
p1=plot(f1(x),a,b,color='green',legend_label='$y=\\frac{df}{dx}$')
p2=plot(f2(x),a,b,color='black',legend_label='$y=\\frac{d^2f}{dx^2}$')
show(p+p1+p2)#,figsize=5)
```

**Problem:**

$$
f(x)=\left\{
  \begin{array}{ll}
    xsin\frac{1}{x} &x&\!\!\!\neq0\\
    0  \qquad &x&\!\!\!=0)
  \end{array}
\right.\\
g(x)=\left\{
  \begin{array}{ll}
    x^2sin\frac{1}{x} &x&\!\!\!\neq 0\\

  \end{array}
\right.
$$

```python
def f(x):
    if(x!=0):
       return x*sin(1/x)
    else:
       return 0

def g(x):
    if(x!=0):
       return x^2*sin(1/x)
    else:
        return 0
f(x)=x*sin(1/x)    
g(x)=x^2*sin(1/x)
p1=plot(f(x),(x,-0.1,0.1),thickness=0.4,color='red',figsize=4,title='$x \\sin 1/x$')
p2=plot(g(x),(x,-0.1,0.1),thickness=0.4,color='green',figsize=4,title='$x^2 \\sin 1/x$')
graphics_array([p1,p2])
```

### Implicit Derivative
Example: Find the derivative $\frac{dy}{dx}$ of
$$
2(x^2+y^2)^2=25(x^2-y^2)
$$

First let us plot the graph

```python
var('x,y')
f(x,y)=2*(x^2+y^2)^2-25*(x^2-y^2)
curve=implicit_plot(f(x,y),(x,-4,4),(y,-2,2))
curve.show(figsize=4)
```

```python
dyx=f.implicit_derivative(y,x)
show(dyx)
```

```python
a,b=3,1
m=dyx.subs(x=a,y=b)
T=b+m*(x-a)
```

```python
pt=point((a,b),size=20,color='black')
tgt=plot(T,2,4,color='red')
show(curve+pt+tgt,figsize=4)
```

Problem: Find points at which tangent is parralel to the x-axis.
$$
y(y^2-1)(y-2)=x(x-1)(x-2)
$$

```python
f(x,y)=y*(y^2-1)*(y-2)-x*(x-1)*(x-2)
curve=implicit_plot(f(x,y),(x,-2,4),(y,-3,3))
show(curve,figsize=4)
```

```python
dyx=f.implicit_derivative(y,x)
show(dyx)
```

```python
nr=dyx.numerator()
xs=solve(nr==0,x,solution_dict=True);xs
```

```python
a, b = xs[0][x], xs[1][x]
a, b
```

```python
y1=f.subs(x=a)(x,y)
r1=y1.roots()
r1
```

```python
y2=f.subs(x=b)(x,y)
r2=y2.roots()
r2
```

```python
pts1=[(a,r1[i][0]) for i in range(len(r1))]
pts2=[(b,r2[i][0]) for i in range(len(r2))]
```

```python
plot_pts1=point2d(pts1,size=20,color='red')
plot_pts2=point2d(pts2,size=20,color='black')
show(curve+plot_pts1+plot_pts2)
```

### Taylor's Theorem
**Theorem:** Let $f$ be $n+1$ times continously differentiable in an interval $I$  containing $a$ and $x$. Then $f(x)=T_n(x)+R_n(x)$, where
$$
T_n(x) = f(a)+f'(a)(x-a)+\frac{f''(a)}{2!}(x-a)^2+\cdots+\frac{f^{(n)}(a)}{n!}(x-a)^n
$$
known as $n$-th degree Taylor's polynomial and 
$$
R_n(x)=\frac{f^{(n+1)}(c)}{(n+1)!}(x-a)^{n+1}
$$
for some $c$ between $a$ and $x$.


```python
f(x)=x*exp(-x^2/2)+sin(3*x)
a,n=0,10
tn(x)=f.taylor(x,a,n)
show(tn(x))
```

```python
a,n=0,5
tn(x)=f.taylor(x,a,n)
pf=f.plot(-2,2,figsize=4)
ptn=tn.plot(-2,2,color='red')
show(pf+ptn,figsize=4,ymax=10,ymin=-10)
```
