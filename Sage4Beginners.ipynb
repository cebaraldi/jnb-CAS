### Sums

#### Sum of a series (Limit of partial sums)
$\sum_{n=1}^{oo}s_n := \lim_{N \to \infty} \sum_{n=1}^{N}s_n$

```python
s(n)=1/(n^2+1)
S=sum(s(n),n,1,oo)
show("\\sum_{n=1}^{\\infty}"+latex(s(n))+"="+latex(S)+"\\approx"+latex(S.n()))
```

### Integration
#### Indefinite Integral
$$
\begin{array}{c}
f(x,y) = -\frac{sin(sin(x))\cdot cos(x)}{x+1}-\frac{cos(sin(x))}{(x+1)^2} \\
\int f(x) dx = ?
\end{array}
$$

```python
f(x) = -sin(sin(x))*cos(x)/(x+1)-cos(sin(x))/(x+1)^2
show("f:"+latex(f))
```

#### Anti-derivative of f

```python
F = f.integrate(x)
show("\\int f\\; dx ="+latex(F(x)))
```

#### Definite Integral

```python
I = f.integrate(x,pi/4,pi/2)
show("\\int_{\\pi/4}^{\\pi/2} f\\; dx ="+latex(I)+"\\approx"+latex(I.n()))
```

Better method is to compute the definte integral numerically from start! It is numerically more stable, more reliable, approximation is better. The classnumerically of functions that can be symbolically integrated is far more narrow than class of functions that we can integrate numerically.

```python
numerical_integral(f,pi/4,pi/2)
```
